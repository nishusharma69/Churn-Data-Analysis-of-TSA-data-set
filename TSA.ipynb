```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv',encoding='unicode_escape')
```

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0") 
df["TotalCharges"]=df["TotalCharges"].astype("float")
#we have coverted the data type from object to float
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

#we have to write a function to convert the data 0,1 to yes or no

```python
df.head(10)
```

```python
ax=sns.countplot(x='Churn',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    plt.title("Count of Customers by churn")
plt.figure(figsize=(4,2))
```

```python
gb=df.groupby(['Churn']).agg({"Churn":"count"})
gb
sns.set(rc={"figure.figsize":(12,6)})
plt.pie(gb["Churn"],labels=gb.index,autopct="%1.1f%%")
plt.title("Persentage of Churned customers")
```

#From the give pie chart we can conclude that 26.5% of the customers are churned out

```python
ax=sns.countplot(x='gender',data=df,hue='Churn')
for bars in ax.containers:
    ax.bar_label(bars)
    plt.figure(figsize=(0.3,0.3))
```

```python
sns.countplot(x='SeniorCitizen',data=df,hue='Churn')
plt.figure(figsize=(0.3,0.3))
plt.title('Churn by Senior Citizens')
```

```python
sns.histplot(x="tenure",data=df,bins=72,hue="Churn",palette=["#FF8C00", "#00008B"])
plt.figure(figsize=(4,3))
plt.title('Churned out on the behalf of tenure')
```

#The people who use the services for longer period has stayed and those who used for 1 or 2 month have churned out

```python
features = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the matplotlib figure
n = len(features)
plt.figure(figsize=(12, 4 * n))

# Create subplots for each feature
for i, feature in enumerate(features):
    plt.subplot(n, 1, i + 1)
    sns.countplot(data=df, x=feature,hue='Churn')
    plt.title(f'Churn by {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
